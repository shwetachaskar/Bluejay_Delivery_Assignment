# Write a program that:
##1. Take the file as an input
##2. Programmatically analyze the file and print in console the name and position of employees
### a)   who has worked for 7 consecutive days.
### b)   who have less than 10 hours of time between shifts but greater than 1 hour.
### c)   Who has worked for more than 14 hours in a single shift.

```python
import csv

# Analyzes an employee file and prints the name and position of each employee.
def analyze_employee_file(file_path):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)

        for row in reader:
            name = row.get('Employee Name', '')
            position = row.get('Position Status', '')

            print(f"Name: {name}, Position: {position}")


file_path = '/content/Assignment_Timecard.xlsx - Sheet1.csv'
analyze_employee_file(file_path)
```

# who has worked for 7 consecutive days.

```python
from datetime import datetime, timedelta

# Find employees who have worked for a specified number of consecutive days.

def find_employees_with_consecutive_days(file_path, consecutive_days=7):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)

        # Dictionary to store the last working date for each employee
        last_working_date = {}
        # List to store names of employees with 7 consecutive days
        consecutive_employees = []

        for row in reader:
            name = row.get('Employee Name', '')
            date_str = row.get('Pay Cycle Start Date', '')

            if name and date_str:
                date = datetime.strptime(date_str, '%m/%d/%Y').date()

                if name not in last_working_date:
                    last_working_date[name] = date
                else:
                    if (date - last_working_date[name]).days == 1:
                        consecutive_days -= 1
                    else:
                        consecutive_days = 7

                    if consecutive_days == 0:
                        consecutive_employees.append(name)
                        consecutive_days = 7

                    last_working_date[name] = date

        # Print names of employees with 7 consecutive days
        if consecutive_employees:
            print("Employees who have worked for 7 consecutive days:")
            for employee in consecutive_employees:
                print(employee)
        else:
            print("No employees found with 7 consecutive days.")

file_path = '/content/Assignment_Timecard.xlsx - Sheet1.csv'
find_employees_with_consecutive_days(file_path)
```

# who have less than 10 hours of time between shifts but greater than 1 hour

```python
import csv
from datetime import datetime, timedelta

#Find employees who have breaks between shifts within a specified time range

def find_employees_with_short_breaks(file_path, min_break=1, max_break=10):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)

        # Dictionary to store the last check-out time for each employee
        last_check_out = {}

        for row in reader:
            name = row.get('Employee Name', '')
            time_out_str = row.get('Time Out', '')

            if name and time_out_str:
                # Use '%m/%d/%Y %I:%M %p' as the format for '09/12/2023 01:53 PM'
                time_out = datetime.strptime(time_out_str, '%m/%d/%Y %I:%M %p').time()

                if name in last_check_out:
                    time_difference = (datetime.combine(datetime.min, time_out) -
                                       datetime.combine(datetime.min, last_check_out[name])).seconds / 3600

                    if min_break < time_difference < max_break:
                        print(f"{name} has a break of {time_difference:.2f} hours between shifts.")

                last_check_out[name] = time_out


file_path = '/content/Assignment_Timecard.xlsx - Sheet1.csv'
find_employees_with_short_breaks(file_path)
```

# Who has worked for more than 14 hours in a single shift

```python
import csv
from datetime import datetime

#Find employees who have worked for more than a specified number of hours in a single shift

def find_employees_long_shifts(file_path, max_hours=14):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)

        for row in reader:
            name = row.get('Employee Name', '')
            time_in_str = row.get('Time', '')
            time_out_str = row.get('Time Out', '')

            if name and time_in_str and time_out_str:
                # Use '%m/%d/%Y %I:%M %p' as the format for '09/12/2023 10:08 AM'
                time_in = datetime.strptime(time_in_str, '%m/%d/%Y %I:%M %p')
                time_out = datetime.strptime(time_out_str, '%m/%d/%Y %I:%M %p')

                shift_duration = (time_out - time_in).seconds / 3600

                if shift_duration > max_hours:
                    print(f"{name} has worked for {shift_duration:.2f} hours in a single shift.")


file_path = '/content/Assignment_Timecard.xlsx - Sheet1.csv'
find_employees_long_shifts(file_path)
```
